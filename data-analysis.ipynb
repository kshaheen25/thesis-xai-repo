```python
# Load the dataset
file_path = "/kaggle/input/xai-dataset-2/clinical_mastitis_cows.csv"  # Update with the correct path
df = pd.read_csv(file_path)
df
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Select numerical columns for heatmap
numerical_cols = df.select_dtypes(include=['float64', 'int64']).columns

# Create the heatmap
plt.figure(figsize=(18, 10))
sns.heatmap(df[numerical_cols].corr(), annot=True, cmap='coolwarm', linewidths=.5)
plt.title('Correlation Heatmap of Clinical Mastitis Data')
plt.show()
plt.savefig('data_heat_map.png')
```
